````python
x = [10, 20, 30]
y = [20, 40, 10]

x2 = [10, 20, 30]
y2 = [40, 10, 30]

plt.plot(x, y, label = "line1-width-3")
plt.plot(x2, y2, label = "line2-width-5")

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.legend()
plt.title('Two lines')
plt.show()
````

````python
x = [10, 20, 30]
y = [20, 40, 10]

x2 = [10, 20, 30]
y2 = [40, 10, 30]

plt.plot(x, y, linestyle = 'dotted', color = 'blue', label = "line1-dotted")
plt.plot(x2, y2, linestyle = 'dashed', color = 'red', label = "line2-dashed")

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.legend()
plt.title('Two lines')
plt.show()
````

````python
x = [1, 4, 5, 6, 7]
y = [2, 6, 3, 6, 3]

plt.plot(x, y, linestyle = 'dotted', color = 'red', marker='o', markerfacecolor='blue', markersize=12)


plt.xlabel('x axis')
plt.ylabel('y axis')

plt.title('Display marker')
plt.show()
````

### P2. Create a file with the following content: 

```
test.txt
1 2
2 4
3 1
```

````python
import csv

x = [1, 2, 3]
y = [2, 4, 1]

with open('test.txt', 'r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))
print(x,y)      
plt.plot(x, y, label = 'loaded from file')
plt.show()
````

### P4. Create the following pie chart.

````python
slices = [31.3, 24.8, 12.4, 11.3, 10.8, 9.4]
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
cols = ['blue', 'orange', 'green', 'red', 'purple', 'brown']

plt.pie(slices, labels = languages, colors = cols, startangle = 90, shadow = True, explode = (0.1, 0, 0, 0, 0, 0), autopct = "%1.2f%%") 

plt.title('Pie chart')

plt.show()


````

### P5. Write a Python program to draw a scatter graph taking a random distribution in X and Y and plotted against each other. You should get something like this.

````python
import numpy as np
import matplotlib.pyplot as plt


N = 250
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)

plt.xlabel('X')
plt.ylabel('Y')

plt.scatter(x, y, c=colors)
plt.show()
````
