### P1. Write a Python code to output the following graph. (or a similar one)
```
The productivity values for the 3 groups are the following: 

g1 = [80, 60, 80, 67, 34]
g2 = [45, 67, 23, 89, 10]
g3 = [30, 90, 50, 20, 90]
```

````python
from pandas import DataFrame
import matplotlib.pyplot as plt
import numpy as np

a=np.array([[80, 60, 80, 67, 34],[45, 67, 23, 89, 10],[30, 90, 50, 20, 90]])
df=DataFrame(a, columns=['M','T','W','T','F'], index=['Group1','Group2','Group3'])

df.T.plot(kind='bar')
````
